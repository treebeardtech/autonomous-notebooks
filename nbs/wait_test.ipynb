# Wait-tool demo notebook.
# Exercises three shapes of long-running work so the agent can try:
#   - `wait(..., timeout=N)` and watch idle time stay low (healthy)
#   - wait and watch idle time climb (silent / hung-looking)
#   - wait and see the job finish before the timeout
#
# Usage (from the agent):
#   exec_all(nbs/wait_test.ipynb)
#   wait(nbs/wait_test.ipynb, timeout=5)   # should still be running
#   wait(nbs/wait_test.ipynb, timeout=60)  # should finish

```python
# Cell 1 — "training loop": steady progress output every ~1s for ~15s.
# While this runs, `wait` should report low idle time (last output < 1s ago).
import time
for step in range(15):
    print(f"step {step + 1}/15  loss={1.0 / (step + 1):.4f}", flush=True)
    time.sleep(1)
print("training done")
```

```python
# Cell 2 — "silent compute": one print up front, then 20s of nothing.
# While this runs, idle time in `wait` status should climb steadily —
# this is what a hung cell would also look like, so the agent has to
# judge by elapsed time and task shape.
import time
print("starting silent work (20s)...", flush=True)
time.sleep(20)
print("silent work done")
```

```python
# Cell 3 — fast finisher, so the job wraps up cleanly after the slow cells.
print("all done")
```
